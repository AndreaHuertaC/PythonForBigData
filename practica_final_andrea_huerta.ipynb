```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
suma = lambda x, y: x+ y 
total = reduce(suma, map(lambda x: articulos[x], compra), 0)
print(total)


# B. Una función que calcule el precio total de la compra con un descuento del 10%
descuento = 0.1
suma_descuento = lambda x, y: round((x+y)*(1 - descuento), 2)
total_descuento = reduce(suma_descuento, map(lambda x: articulos[x], compra), 0)
print(total_descuento)

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
suma_descuento_condicional = lambda x, y: round((x+y)*(1 - descuento), 2) if total > 100 else x + y 
total_descuento_condicional = reduce(suma_descuento_condicional, map(lambda x: articulos[x], compra), 0)
print(total_descuento_condicional)

# D. Una función que calcule el precio total aplicando el IVA (21%)
iva = 0.21
suma_iva = lambda x, y: round((x + y) * (1 + iva), 2)
total_iva = total = reduce(suma_iva, map(lambda x: articulos[x], compra), 0)
print(total_iva)

# E.lista los artículos cuyo precio es superior a 20€
consolidado = zip(articulos.keys(), articulos.values())
print(list(map(lambda y: y[0], filter(lambda x: x[1] > 20, consolidado))))
```

```python
list_alumnos = lambda curso: map(lambda x: (x[0], x[1], reduce(lambda x, y: x + y, x[3], 0)/len(x[3])), filter(lambda x: x[2] == curso, alumnos))

list(list_alumnos('2'))


```

```python
def lista_alumnos(curso):
    return [(alumno[0], alumno[1], sum(alumno[3])/len(alumno[3])) for alumno in alumnos if alumno[2] == curso]

lista_alumnos('2')
```
